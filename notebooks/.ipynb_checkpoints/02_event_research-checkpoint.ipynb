```python
# Task 1: Event Research for Brent Oil Prices
print("Starting event research...")
print("We need to identify 10-15 key events that affected oil prices")
```

```python
# List of events to research (we'll fill in dates)
events_to_research = [
    "2008 Financial Crisis",
    "Arab Spring (2010-2012)",
    "OPEC production cuts 2014",
    "US shale oil boom",
    "COVID-19 pandemic",
    "Russia-Ukraine war 2022",
    "2011 Libya civil war",
    "2014 Oil price crash",
    "2019 drone attacks on Saudi Arabia",
    "2020 Negative oil prices",
    "Iran sanctions",
    "Venezuela economic crisis",
    "Iraq War 2003",
    "1990 Gulf War",
    "2015 Iran nuclear deal"
]

print("Events to research:", len(events_to_research))
for i, event in enumerate(events_to_research, 1):
    print(f"{i}. {event}")
```

```python
# Add approximate dates to events
import pandas as pd

events_data = [
    {"Event": "2008 Financial Crisis", "Date": "2008-09-15", "Type": "Economic", "Description": "Lehman Brothers collapse triggered global recession"},
    {"Event": "Arab Spring", "Date": "2010-12-17", "Type": "Political", "Description": "Series of anti-government protests across Middle East"},
    {"Event": "OPEC production cuts 2014", "Date": "2014-11-27", "Type": "Policy", "Description": "OPEC decided not to cut production, leading to price crash"},
    {"Event": "COVID-19 pandemic", "Date": "2020-03-11", "Type": "Economic", "Description": "WHO declared pandemic, global lockdowns reduced oil demand"},
    {"Event": "Russia-Ukraine war 2022", "Date": "2022-02-24", "Type": "Political", "Description": "Russian invasion led to sanctions and supply concerns"},
    {"Event": "2011 Libya civil war", "Date": "2011-02-15", "Type": "Conflict", "Description": "Civil war disrupted Libyan oil exports"},
    {"Event": "2014 Oil price crash", "Date": "2014-06-20", "Type": "Market", "Description": "Prices fell from $100 to $50 due to oversupply"},
    {"Event": "2019 drone attacks on Saudi Arabia", "Date": "2019-09-14", "Type": "Conflict", "Description": "Attack on Saudi oil facilities disrupted 5% of global supply"},
    {"Event": "2020 Negative oil prices", "Date": "2020-04-20", "Type": "Market", "Description": "WTI crude futures fell to -$37 due to storage shortage"},
    {"Event": "Iran sanctions", "Date": "2018-05-08", "Type": "Political", "Description": "US reinstated sanctions on Iranian oil exports"},
    {"Event": "Iraq War 2003", "Date": "2003-03-20", "Type": "Conflict", "Description": "US invasion of Iraq disrupted Middle East oil production"},
    {"Event": "1990 Gulf War", "Date": "1990-08-02", "Type": "Conflict", "Description": "Iraq's invasion of Kuwait led to oil price spike"},
    {"Event": "2015 Iran nuclear deal", "Date": "2015-07-14", "Type": "Political", "Description": "Deal lifted sanctions, allowing Iranian oil back to market"}
]

# Create DataFrame
events_df = pd.DataFrame(events_data)
events_df['Date'] = pd.to_datetime(events_df['Date'])

print("Events DataFrame created:")
print(events_df.head())
print(f"\nTotal events: {len(events_df)}")
```

```python
# Save events to CSV
events_df.to_csv('../data/key_events.csv', index=False)
print("Events saved to: ../data/key_events.csv")
print("File preview:")
print(events_df[['Event', 'Date']].head())
```
